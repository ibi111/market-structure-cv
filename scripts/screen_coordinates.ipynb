```python
import pyautogui
import time

print("Move mouse to TOP-LEFT of chart in 3 seconds...")
time.sleep(10)
x1, y1 = pyautogui.position()
print(f"Top-left: ({x1}, {y1})")

print("Move mouse to BOTTOM-RIGHT of chart in 3 seconds...")
time.sleep(5)
x2, y2 = pyautogui.position()
print(f"Bottom-right: ({x2}, {y2})")

print(f"\nRegion: ({x1}, {y1}, {x2}, {y2})")
```
